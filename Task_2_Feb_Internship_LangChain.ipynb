STEP 5: Create Embeddings (Gemini)

```python
from langchain_community.embeddings import HuggingFaceEmbeddings

embeddings = HuggingFaceEmbeddings(
    model_name="all-MiniLM-L6-v2"
)

print(" Embeddings working!")
```

STEP 3: Load PDF

```python
from langchain_community.document_loaders import PyPDFLoader

loader = PyPDFLoader("/content/New Microsoft Word Document.pdf")
documents = loader.load()

print("PDF Loaded Successfully")
```

STEP 4: Split Text into Chunks

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=50
)

docs = text_splitter.split_documents(documents)

print(" Text Split Done")
```

STEP 6: Create Vector Store (FAISS)

```python
from langchain_community.vectorstores import FAISS

vectorstore = FAISS.from_documents(docs, embeddings)

print(" Vector Store Created")
```

STEP 7: Create Retriever

```python
retriever = vectorstore.as_retriever()
```

🤖 STEP 8: Create Gemini LLM

```python
from langchain_google_genai import ChatGoogleGenerativeAI

llm = ChatGoogleGenerativeAI(
    model="gemini-pro",
    google_api_key=api_key
)
```

STEP 9: Create QA Chain

```python
from langchain.chains import RetrievalQA

qa_chain = RetrievalQA.from_chain_type(
    llm=llm,
    retriever=retriever
)

print(" QA Chain Ready")
```

STEP 10: Ask Questions (Main Loop)

```python
while True:
    query = input("\nAsk a question (type 'exit' to quit): ")

    if query.lower() == "exit":
        break

    answer = qa_chain.run(query)
    print(" Answer:", answer)
```
